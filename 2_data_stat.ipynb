```python
import pandas as pd

# 파일 경로
train_path = "２_train_user_dataset.tsv"
dev_path = "２_dev_user_dataset.tsv"

# 데이터 불러오기
train_df = pd.read_csv(train_path, sep="\t", header=None)
dev_df = pd.read_csv(dev_path, sep="\t", header=None)

train_behaviors = pd.read_csv('download/MINDsmall_train/behaviors.tsv', sep='\t',
                              names=['ImpressionID', 'UserID', 'Time', 'History', 'Impressions'])
test_behaviors = pd.read_csv('download/MINDsmall_dev/behaviors.tsv', sep='\t',
                             names=['ImpressionID', 'UserID', 'Time', 'History', 'Impressions'])

behaviors_all = pd.concat([train_behaviors, test_behaviors]).reset_index(drop=True)

# 공통 통계 계산 함수
def dataset_stats_user_mode(df):
    """user_dataset (train/dev) 구조용 통계"""
    num_items = df[1].nunique()
    users = set()
    clicks = 0
    for row in df[3]:
        for token in row.split():
            users.add(token.split("-")[0])
            if token.endswith("-1"):
                clicks += 1
    return {"users": len(users), "items": num_items, "clicks": clicks}

def dataset_stats_behaviors(df):
    """behaviors.tsv 구조용 통계"""
    users = df["UserID"].nunique()
    items = set()
    clicks = 0
    for row in df["Impressions"].dropna():
        for token in row.split():
            items.add(token.split("-")[0])
            if token.endswith("-1"):
                clicks += 1
    return {"users": users, "items": len(items), "clicks": clicks}

# 개별 통계
train_stats = dataset_stats_user_mode(train_df)
dev_stats = dataset_stats_user_mode(dev_df)
combined_stats = dataset_stats_user_mode(pd.concat([train_df, dev_df]))
beh_stats = dataset_stats_behaviors(behaviors_all)

# 정리
datasets = [("User Dataset", combined_stats), ("Behaviors", beh_stats)]

def calc_sparsity(stats):
    return 1 - (stats["clicks"] / (stats["users"] * stats["items"]))

# 📌 결과 출력 (print)
print(f"{'Dataset':<12}{'# of users':>12}{'# of items':>12}{'# of clicks':>15}{'Sparsity':>12}")
for name, stats in datasets:
    sparsity = calc_sparsity(stats)
    print(f"{name:<12}{stats['users']:>12,}{stats['items']:>12,}{stats['clicks']:>15,}{sparsity*100:>14.2f}%")
```
